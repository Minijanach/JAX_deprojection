```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def xi(x,k,d):
    return x*k +d
def p_q_change(x):
    return 1 / (1 + np.exp(-np.abs(x)))
```

```python
def plot_triaxial_ellipsoid_const(p1, e1):
    
    a = 2.0  
    b = 1  
    c = 1  #

    # Define the grid of points
    x_range = np.linspace(-a, a, 100)
    y_range = np.linspace(-b, b, 100)
    z = 0

    x_grid, y_grid = np.meshgrid(x_range, y_range)


    p = p1
    q = 1 #c/a
    e= e1
    
    R = ((np.abs(x_grid)**(2 - e) + (np.abs(y_grid) / p)**(2 - e) + (np.abs(z) / q)**(2 - e)))**(1/(2-e))
    

    # Create the contour plot
    #plt.figure(figsize=(8, 6))
    contour = plt.contour(x_grid, y_grid, R, levels=10, colors='b')
    #plt.clabel(contour, inline=1, fontsize=10)
    #plt.title('Contour Plot of Triaxial Ellipsoid')
    #plt.xlabel('X-axis')
    #plt.ylabel('Y-axis')
    plt.gca().set_aspect('equal', adjustable='box')
    #plt.show()
```

```python
plot_triaxial_ellipsoid_const(1, 0)
plot_triaxial_ellipsoid_const(1, -1)
plot_triaxial_ellipsoid_const(1, 1)
plot_triaxial_ellipsoid_const(0.8, 0)
plot_triaxial_ellipsoid_const(0.8, -1)
plot_triaxial_ellipsoid_const(0.8, 1)
plt.show()
```
